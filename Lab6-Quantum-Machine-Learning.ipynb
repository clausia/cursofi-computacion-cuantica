## 1.A Perceptrón Clásico

Un **perceptrón** es una unidad básica de una red neuronal artificial, que se utiliza en el aprendizaje automático para tareas de clasificación. Es el modelo más simple de una red neuronal, diseñado por [Frank Rosenblatt](https://en.wikipedia.org/wiki/Frank_Rosenblatt) en 1957.

1. **Entradas (Inputs)**: 
   - El perceptrón recibe múltiples entradas, denotadas como $i_1, i_2, ..., i_n$.
   
2. **Pesos (Weights)**:
   - Cada entrada $i_j$ tiene un peso asociado $w_j$ que determina la importancia de esa entrada en la predicción final.

3. **Suma Ponderada (Weighted Sum)**:
   - El perceptrón calcula una suma ponderada de las entradas: 
     $$ z = \sum_{j=1}^{n} w_i i_j + b $$
   - Aquí, $b$ es el sesgo (bias), que ajusta la salida de la suma ponderada.

4. **Función de Activación (Activation Function)**:
   - La suma ponderada $z$ se pasa a través de una función de activación para producir la salida $y$. En un perceptrón simple, esta es a menudo una función escalón:
$$ y = f(z) = \begin{cases} 
 1 & \text{si } z \geq 0 \\
 0 & \text{si } z < 0 
\end{cases} $$

El perceptrón se entrena ajustando los pesos y el sesgo para minimizar el error en las predicciones. Esto se realiza mediante un algoritmo de optimización, como el gradiente descendiente, que ajusta los pesos y el sesgo iterativamente para mejorar el rendimiento del modelo.

## 1.B Perceptrón Cuántico

En los últimos años se han presentado varias propuestas para implementar perceptrones en computadoras cuánticas. El concepto más investigado es el de "_qubit neurona_", en la que _cada qubit actúa como una neurona individual_ dentro de la red. La mayor parte del esfuerzo de investigación se ha dedicado a explotar la no linealidad del proceso de medición para implementar la función de umbral.

$$ $$

Aquí veremos el diseño mostrado en el trabajo de [Tacchino et.al.](https://www.nature.com/articles/s41534-019-0140-4) que imita fielmente un perceptrón de Rosenblatt de valor binario en una computadora cuántica. 

Primero, el equivalente de los vectores de peso y entrada clásicos de $m$ dimensiones se codifica en un circuito cuántico utilizando $N$ qubits, donde $m = 2^N$. Por un lado, esto evidentemente permite explotar la ventaja exponencial del almacenamiento de información cuántica, como ya se ha señalado.

El trabajo de [Tacchino et.al.](https://www.nature.com/articles/s41534-019-0140-4) presenta un procedimiento original para generar estados entrelazados multipartitos basados en principios de información cuántica que permite reducir de manera crucial los recursos computacionales cuánticos a emplear. Pero aquí veremos solamente la versión de "fuerza bruta" por simplicidad.

Este modelo de perceptrón cuántico se puede utilizar para clasificar patrones simples, como líneas verticales u horizontales, entre todas las entradas posibles.

Los vectores de entrada y pesos estarán limitados a valores binarios, $i_j, w_j \in \{−1, 1\}$, como en las [neuronas de McCulloch-Pitts](https://es.wikipedia.org/wiki/Neurona_de_McCulloch-Pitts). Por lo tanto, un vector de entrada $m$-dimensional se codifica utilizando los $m$ coeficientes necesarios para definir una función de onda general $\ket{\psi_i}$ de $N$ qubits.

Dados los vectores de entreda ($\vec{i}$) y pesos ($\vec{w}$):

$$\vec{i}=\pmatrix{i_0 \\ i_1 \\ \vdots \\ i_{m-1}} ~~~~~~~~~,~~~~~~~~ \vec{w}=\pmatrix{w_0 \\ w_1 \\ \vdots \\ w_{m-1}}$$

Así, definimos los dos estados cuánticos:

$$\ket{\psi_i}\frac{1}{\sqrt{m}}\sum_{j=0}^{m-1}i_j\ket{j} ~~~~~~~~~,~~~~~~~~ \ket{\psi_w}\frac{1}{\sqrt{m}}\sum_{j=0}^{m-1}w_j\ket{j}$$

$$ $$

Los estados $\ket{j}\in\ket{00\cdots 00},\ket{00\cdots 01},\cdots,\ket{11\cdots 11}$ froman la base computacional de $N$ qubits, y corresponden a todos los posibles estados. Como es usual, estos estados se etiquetan con númreos enteros $j\in\{0,\cdots,m-1\}$.

### Paso 1

Preparar el estado $\ket{\psi_i}$ codificando los valores de entrada del vector $\vec{i}$. 

Los qubits comienzan en el estado $\ket{00\cdots 00}\equiv\ket{0}^{\otimes N}$, entonces se aplica una transformación unitaria $U_i$, tal que 

$$U_i\ket{0}^{\otimes N}=\ket{\psi_i}$$

### Paso 2

Calcular el producto interno entre $\vec{w}$ y $\vec{i}$ usando el registro cuántico.

Esta tarea puede ser ejecutada eficientemente al definir una transformación $U_w$ tal que el estado cuántico de pesos sea rotado como

$$U_w\ket{\psi_w}=\ket{1}^{\otimes N}=\ket{m-1}$$

Si aplicamos $U_w$ después de $U_i$, el estado cuántico de $N$ qubits se convierte en

$$U_w\ket{\psi_i}=\sum_{j=0}^{m-1}c_j\ket{j}\equiv\ket{\phi_{i,w}}$$

Notemos que el producto escalar entre los dos estados cuánticos es

\begin{align*}
	\langle \psi_w|\psi_i\rangle&=\langle \psi_w | U_w^{\dagger}U_w | \psi_i\rangle\\
	&=\langle m-1 | \phi_{i,w} \rangle\\
	&=c_{m-1}\\
	\end{align*}

- $U_w$ rota el estado $\ket{\psi_w}$ de modo que se alinee con el estado $\ket{11\cdots 11}=\ket{m-1}$
- $U_w$ rota el estado $\ket{\psi_i}$ con los mismos ángulos
- Después de aplicar $U_w$ a $\ket{\psi_i}$, el producto interno corresponde a la proyección de $U_w\ket{\psi_i}$ sobre el eje del estado $\ket{11\cdots 11}$

De este modo, dada la definición de los estados cuánticos $\ket{\psi_i}$ y $\ket{\psi_w}$, vemos que el producto escalar de los vectores de entrada y pesos es 

$$\vec{w}\cdot\vec{i}=m\langle \psi_w|\psi_i\rangle$$

Así, el resultado deseado está contenido, hasta un factor de normalización, en el coeficiente $c_{m-1}$ del estado final $\phi_{i,w}$

### Paso 3

Para extraer dicha información, usamos un qubit ancilla ($a$) inicialmente establecido en el estado $\ket{0}$. Una compuerta $NOT$ multicontrolada entre los $N$ qubits como controles y el qubit $a$ como objetivo, esto conduce a

$$ \ket{\phi_{i,w}}\ket{0}_a\to\sum_{j=0}^{m-2}c_j\ket{j}\ket{0}_a+c_{m-1}\ket{m-1}\ket{1}_a$$

### Medición

La no linealidad requerida por la función umbral en la salida del perceptrón se obtiene inmediatamente realizando una medición cuántica: de hecho, al medir el estado del qubit ancilla en la base computacional se produce la salida $\ket{1}_a$ (es decir, un perceptrón activado) con probabilidad 

$$|c_{m-1}|^2$$

### Implementación de las transformaciones unitarias

Definimos un bloque de cambio de signo (sign flip), $SF_{N,j}$, como la transformación unitaria que actúa sobre la base computacional de $N$ qubits de la siguiente manera:

$$ SF_{N,j}\ket{j'} = \begin{cases} 
 \ket{j'} & \text{si } j\neq j' \\
 -\ket{j'} & \text{si } j = j' 
\end{cases} $$

En general, $SF_{N,j}$ es equivalente a una operación cuántica multicontrolada entre $N$ qubits: por ejemplo, para cualquier $N$, $m = 2^N$, una compuerta $Z$ controlada entre $N$ qubits ($C^NZ$) es una compuerta cuántica bien conocida que realiza $SF_{N, m−1}$, mientras que una compuerta de un qubit $Z$ actúa como $SF_{1,1}$. Para una entrada dada $\vec{i}$, la unitaria $U_i$ se puede obtener combinando rotaciones simples de qubits únicos y bloques de cambio de signo para introducir los factores $−1$ requeridos delante de los estados básicos $\ket{j}$ en la representación del estado objetivo $\ket{\psi_i}$.

Se puede implementar toda la familia de bloques $SF_{N,j}$ para $N$ qubits utilizando compuertas $C^NZ$ en combinación con compuertas $NOT$ de un solo qubit (es decir, operaciones de cambio de un solo bit):

$$ SF_{N,j}=O_j\left(C^NZ\right)O_j$$

donde

$$O_j=\otimes_{l=0}^{m-1}\left(NOT_l\right)^{1-j_l}$$

En la expresión anterior, $NOT_l$ significa que el cambio de bit se aplica al $l$-ésimo qubit y $j_l = 0$ ($1$) si el $l$-ésimo qubit está en el estado $\ket{0}$ $(\ket{1})$ en el estado de la base computacional $\ket{j}$.

$$ $$

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import partial_trace, Statevector
from qiskit.circuit.library import ZGate, XGate
```

```python
def createSFblock(N, j):
    
    circuit = QuantumCircuit(N)
    
    state_decimal = format(j, '0'+str(N)+'b')[::-1] # orden inverso debido al orden de Qiskit: |q1q0>
    
    for qubit in range(N):
        if state_decimal[qubit] == '0':
            circuit.x(qubit)
    
    multiConrolledZ = ZGate().control(num_ctrl_qubits=N-1)
    circuit.append(multiConrolledZ, range(N))
    
    for qubit in range(N):
        if state_decimal[qubit] == '0':
            circuit.x(qubit)
    
    return circuit
```

```python
def U_i(N, ki, circuit):
    
    # crear superposición
    circuit.h(range(N))
    
    # aplicar bloques SF según la información del ki
    vector_i = format(ki, '0'+str(2**N)+'b')
    for ij in range(len(vector_i)):
        if vector_i[ij] == '1': # n_j = 1, equivalente a tener -1 en el vector i
            circuit.compose(createSFblock(N, ij), inplace=True)  # el estado i es el que debe tener la fase de -1

def U_w(N, kw, circuit):
    
    # aplicar bloques SF según la información del kw
    vector_w = format(kw, '0'+str(2**N)+'b')
    for wj in range(len(vector_w)):
        if vector_w[wj] == '1': # n_j = 1,equivalente a tener -1 en el vector w
            circuit.compose(createSFblock(N, wj), inplace=True)  # el estado w es el que debe tener la fase de -1
    
    # aplicar Hadamards
    circuit.h(range(N))
    
    # aplicar NOTs
    circuit.x(range(N))
```

```python
def perceptron(N, ki, kw, draw=False):
    
    # crear circuito, N qubits + ancilla
    circuit = QuantumCircuit(N + 1, 1)
    
    # aplicar U_i
    U_i(N, ki, circuit)
    
    # aplicar U_w
    U_w(N, kw, circuit)
    
    # aplicar C^N X
    circuit.mcx(control_qubits=[i for i in range(N)], target_qubit=N)
    
    # medir el qubit ancilla
    circuit.measure(N, 0)
    
    # dibujar el circuito
    if draw == True:
        display(circuit.draw('mpl'))
    
    return circuit
```

```python
perceptron_circuit = perceptron(N=2, ki=11, kw=7, draw=True) # equivalente al circuito en la figura 3c del paper de Tacchino et. al.
```

```python
from qiskit.primitives import Sampler

sampler = Sampler()
job = sampler.run(perceptron_circuit)

print(job.result().quasi_dists[0].binary_probabilities())        # se observa '0' con el 100% de probabilidad, perceptrón no activado
```

```python
perceptron_circuit = perceptron(N=2, ki=11, kw=11, draw=True)    # notar que i = w
```

```python
job = sampler.run(perceptron_circuit)

print(job.result().quasi_dists[0].binary_probabilities())        # se observa '1' con el 100% de probabilidad, perceptrón activado
```

## 2.A Clasificación

Preparamos un conjunto de datos de clasificación simple, generando $20$ valores muestra, cada uno una tupla $(x_1,x_2)$ y definiendo su resultado $y$ dependiendo de si su suma es mayor a $0$ o no, asignando $\{1, 0\}$ respectivamente.

En la variable `y_one_hot` se guarda la codificación [one-hot](https://en.wikipedia.org/wiki/One-hot).

En la variable `y` se reetiquetan las muestras con valores $\{-1, +1\}$, los cuales son usados para decidir de qué color se mostrará la muestra en la gráfica, los valores con la etiqueta $+1$ son los que están por encima de la línea $x=-y$ y los de etiqueta $-1$ son los que están por debajo de dicha línea.

```python
num_inputs = 2                                             # cantidad de entradas (primera capa de la red neuronal)

num_samples = 20                                           # cantidad de muestras para entrenar la red neuronal

X = 2*np.random.rand(num_samples, num_inputs) - 1          # de manera aleatoria creamos lo valores que entran a la capa 1
                                                           # estos son dos valores (inputs) por cada muestra

y01 = 1*(np.sum(X, axis=1) >= 0)                           # si la suma de los dos valores (de cada muestra) es >= 0
                                                           # (es decir, están arriba de la línea x=-y en el plano)
                                                           # entonces lo clasificamos con la etiqueta '1', caso contrario
                                                           # con la etiqueta '0', así los valores de y01 ∈ {0, 1}
                                                           

y = 2*y01-1                                                # en esta variable reetiquetamos con: y ∈ {-1, +1}

y_one_hot = np.zeros((num_samples, 2))                     # variable para la codificación one-hot

for i in range(num_samples):                               # generamos los valores one-hot de cada muestra usando y01
    y_one_hot[i, y01[i]] = 1                               # si el valor en y01[i] = 0, entonces y_one_hot_i = [1, 0]
                                                           # pero si y01[i] = 1, entonces y_one_hot_i = [0, 1]

    
for x, y_target in zip(X, y):                              # iteramos sobre las muestras y sus respectivos resultados
    if y_target == 1:                                      # si se trata de una muestra con resultado etiquetado con '+1'
        plt.plot(x[0], x[1], 'bo')                         # la dibujamos en azul
    else:                                                  # si la muestra tiene un resultado con la etiqueta '-1'
        plt.plot(x[0], x[1], 'go')                         # entonces la dibujamos en verde


plt.plot([-1, 1], [1, -1], '--', color='black')            # definimos el plot con la línea x = -y
plt.show()                                                 # mostramos el plot
```

### Clasificación con un `EstimatorQNN`

Primero mostramos cómo se puede usar un `EstimatorQNN` (hereda de `NeuralNetwork`) para la clasificación dentro de una `NeuralNetworkClassifier`. En este contexto, se espera que `EstimatorQNN` devuelva una salida unidimensional en $\left[-1,+1\right]$. Esto solo funciona para la clasificación binaria y asignamos las dos clases a $\{-1,+1\}$. Esta clase necesita un [mapa de características](https://computersciencewiki.org/index.php/Feature_maps_(Activation_maps)) y un [ansatz](https://en.wikipedia.org/wiki/Ansatz).

$$ $$

Un **feature map** (mapa de características) es una técnica utilizada para codificar datos clásicos en un estado cuántico. Su objetivo es transformar los datos de entrada en un espacio de características (feature space) más elevado donde los datos pueden ser procesados por algoritmos cuánticos de manera más efectiva.

El `ZZFeatureMap` de Qiskit es un ejemplo específico de un _feature map_. Este mapa de características utiliza una combinación de compuertas Hadamard y rotaciones controladas $ZZ$ para codificar las características de los datos en qubits.

$$ $$

La palabra **ansatz** proviene del alemán y significa "aproximación" o "suposición". En el contexto de la computación cuántica, un _ansatz_ se refiere a una suposición sobre la forma de la solución cuántica que se utiliza para encontrar una solución aproximada a un problema.

Los _circuitos ansatz_ están formados por compuertas cuánticas parametrizadas. Estos parámetros se ajustan durante el proceso de optimización para minimizar una función de costo específica.

El `RealAmplitudes` es un circuito ansatz predefinido en Qiskit, utilizado frecuentemente en algoritmos variacionales y otras aplicaciones de QML. Este tipo de ansatz es conocido por su simplicidad y eficacia en capturar la estructura de los estados cuánticos que se desean aproximar. Consiste en una secuencia de compuertas parametrizadas que aplican rotaciones en el eje $Y$ a cada qubit, seguidas de capas de entrelazamiento entre los qubits.

```python
# construir la QNN
qc = QuantumCircuit(2)

feature_map = ZZFeatureMap(2)    # para mapear las características originales a unas nuevas, unas que sean cuánticas

ansatz = RealAmplitudes(2)       # circuito inicial, se usa una 'educated guess' (ansatz), en particular el llamado RealAmplitudes
                                 # este es el modelo que queremos optimizar
                                 # repeticiones = 3 (valor default)

qc.compose(feature_map, inplace=True)
qc.compose(ansatz, inplace=True)
qc.draw(output="mpl")
```

```python
# veamos el detalle del mapa de características
feature_map.decompose().draw(output="mpl")
```

```python
# veamos el detalle del ansatz
ansatz.decompose().draw(output="mpl")
```

Creamos una red neuronal cuántica:

```python
estimator_qnn = EstimatorQNN(
    circuit=qc, input_params=feature_map.parameters, weight_params=ansatz.parameters
)
```

```python
# la QNN mapea las entradas a [-1, +1]
estimator_qnn.forward(X[0, :], algorithm_globals.random.random(estimator_qnn.num_weights))  

# Args: 
#      input_data: datos de entrada de la forma (num_inputs)
#      weights: pesos entrenables de la forma (num_weights)
# Returns:
#      El resultado de la red neuronal de la forma (output_shape)
```

```python
# función de devolución de llamada (callback) que dibuja un diagrama en tiempo real cuando se llama al método .fit()
def callback_graph(weights, obj_func_eval):
    clear_output(wait=True)
    objective_func_vals.append(obj_func_eval)
    plt.title("Valor de la función objetivo contra a la iteración")
    plt.xlabel("Iteración")
    plt.ylabel("Valor de la función objetivo")
    plt.plot(range(len(objective_func_vals)), objective_func_vals)
    plt.show()
```

### Clasificación con un `SamplerQNN`

Ahora, mostramos cómo se puede usar un `SamplerQNN` (hereda de `NeuralNetwork`) para la clasificación dentro de una `NeuralNetworkClassifier`. En este contexto, se espera que `SamplerQNN` devuelva un vector de probabilidad de dimensión $d$ como salida, donde $d$ denota el número de clases. La primitiva Sampler subyacente devuelve cuasi-distribuciones de cadenas de bits y solo necesitamos definir un mapeo de las cadenas de bits medidas a las diferentes clases. Para la clasificación binaria usamos el mapeo de paridad.

```python
# construir el mapa de características
feature_map = ZZFeatureMap(num_inputs)

# construir el ansatz
ansatz = RealAmplitudes(num_inputs, reps=1)

# construir el circuito cuántico
qc = QuantumCircuit(num_inputs)
qc.append(feature_map, range(num_inputs))
qc.append(ansatz, range(num_inputs))
qc.decompose().draw(output="mpl")
```

```python
# la paridad asigna cadenas de bits a 0 o 1
def parity(x):
    return '{:b}'.format(x).count('1') % 2

output_shape = 2  # corresponde al número de clases, posibles resultados del mapeo (de paridad)
```

```python
# construir la QNN
sampler_qnn = SamplerQNN(
    circuit=qc,
    input_params=feature_map.parameters,
    weight_params=ansatz.parameters,
    interpret=parity,
    output_shape=output_shape,
)
```

```python
# construir el clasificador
sampler_classifier = NeuralNetworkClassifier(
    neural_network=sampler_qnn, optimizer=COBYLA(maxiter=30), callback=callback_graph
)
```

```python
# crear una lista vacía para la devolución de llamada para almacenar evaluaciones de la función objetivo
objective_func_vals = []
plt.rcParams["figure.figsize"] = (12, 6)

# ajustar el clasificador a los datos
sampler_classifier.fit(X, y01)

# regresar al tamaño de figura predeterminado
plt.rcParams["figure.figsize"] = (6, 4)

# score del clasificador
sampler_classifier.score(X, y01)
```

```python
# evaluate data points
y_predict = sampler_classifier.predict(X)                        # obtenemos las predicciones para todas las muestras


# graficar resultados
# rojo => clasificado incorrectamente

for x, y_target, y_p in zip(X, y01, y_predict):                  # iteramos sobre todas las muestras
    if y_target == 1:                                            # si estaba clasificada con '+1'
        plt.plot(x[0], x[1], 'bo')                               # la dibujamos en azul
    else:                                                        # si estaba clasificada con '-1'
        plt.plot(x[0], x[1], 'go')                               # la dibujamos en verde
    if y_target != y_p:                                          # si la predicción es diferente a la etiqueta
                                                                 # colocamos un círculo rojo alrededor
        plt.scatter(x[0], x[1], s=200, facecolors='none', edgecolors='r', linewidths=2)


plt.plot([-1, 1], [1, -1], '--', color='black')                  # definimos el plot con la línea x = -y
plt.show()                                                       # mostramos el plot
```

Nuevamente, una vez que el modelo está entrenado, podemos echar un vistazo a los pesos. Como configuramos `reps=1` explícitamente en el ansatz, tendremos menos parámetros que en el modelo anterior.

### Clasificador Cuántico Variacional (Variational Quantum Classifier) `VQC`

La clase `VQC` es una variante especial del `NeuralNetworkClassifier` con un `SamplerQNN`. Aplica un mapeo de paridad (o extensiones a múltiples clases) para mapear desde la cadena de bits hasta la clasificación, lo que da como resultado un vector de probabilidad, que se interpreta como un resultado codificado con _one-hot_. De forma predeterminada, aplica la función `CrossEntropyLoss` que espera que las etiquetas se proporcionen en formato de codificación _one-hot_ y también devolverá predicciones en ese formato.

```python
# construir el mapa de características, ansatz y optimizador
feature_map = ZZFeatureMap(num_inputs)
ansatz = RealAmplitudes(num_inputs, reps=1)

# construir el clasificador cuántico variacional
vqc = VQC(
    feature_map=feature_map,
    ansatz=ansatz,
    loss="cross_entropy",
    optimizer=COBYLA(maxiter=30),
    callback=callback_graph,
)
```

## 2.B Regresión

Preparamos un conjunto de datos de regresión simple para ilustrar los siguientes algoritmos.

```python
num_samples = 20                                           # cantidad de muestras para entrenar la red neuronal

eps = 0.2                                                  # valor de tolerancia (epsilon)

lb, ub = -np.pi, np.pi                                     # limites inferior y superior

X_ = np.linspace(lb, ub, num=50).reshape(50, 1)            # generar 50 números en el intervalo (lb, ub) para graficar f

f = lambda x: np.sin(x)                                    # función lambda que calcula el seno

X = (ub - lb)*np.random.rand(num_samples, 1) + lb          # generar 20 (num_samples) muestras de forma alaeatoria
                                                           # pero en el intervalo (lb, ub)

y = f(X[:,0]) + eps*(2*np.random.rand(num_samples)-1)      # generar los resultados (vector 'y') aleatoriamente
                                                           # pero siguiendo una función seno, y con tolerancia de 'eps'
                                                           # es decir, ± 0.2 alrededor de la función seno

plt.plot(X_, f(X_), 'r--')                                 # graficar la función f (seno), usando los 50 puntos en _X   
plt.plot(X, y, 'bo')                                       # graficar la muestras
plt.show()                                                 # mostrar gráfica
```
